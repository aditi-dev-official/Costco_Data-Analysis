# Business Requirements

* KPI's Requirements

```python
	
# Total Sales
total_sales = df['Sales'].sum()

# Average Sales
avg_sales = df['Sales'].mean()

# Number of Items Sold (no Quantity column, so fallback = count of rows)
no_of_items_sold = df['Sales'].count()

# Average Rating (column name is 'Rating', not 'Ratings')
avg_ratings = df['Rating'].mean()
```

```python
import matplotlib.pyplot as plt

# 1. Filter the DataFrame to keep only the 2 categories you want
# Assuming your two categories are 'Low Fat' and 'Regular' (excluding 'reg')
df_filtered = df[df['Item Fat Content'].isin(['Low Fat', 'Regular'])]

# 2. Group and sum the filtered data
sales_by_fat = df_filtered.groupby('Item Fat Content')['Sales'].sum()

# 3. Plot the donut chart
plt.pie(
    sales_by_fat,
    labels=sales_by_fat.index,
    autopct='%.1f%%',
    startangle=90,
)

plt.title('Sales by Fat Content')
plt.axis('equal') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group and sort sales by item type
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

# Set figure size
plt.figure(figsize=(10, 6))

# Create bar chart
bars = plt.bar(sales_by_type.index, sales_by_type.values, color='brown')

# Rotate x-axis labels for readability
plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')
# Add value labels on top of each bar
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():,.0f}',
        ha='center',
        va='bottom',
        fontsize=8
    )

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group and Sort (Largest to Smallest)
data = df.groupby('Outlet Location Type')['Sales'].sum().sort_values(ascending=False)

# Calculate centering offsets
offsets = (data.max() - data) / 2

# Plot
plt.barh(data.index, data, left=offsets, color='skyblue')
plt.gca().invert_yaxis()
plt.title('Sales Funnel by Location')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Define funnel_data by grouping your main dataframe (df)
# This sums 'Sales' for each 'Outlet Location Type' and sorts them
funnel_data = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
funnel_data = funnel_data.sort_values('Sales', ascending=False)

# 2. Extract labels and values from the newly defined funnel_data
labels = funnel_data['Outlet Location Type']
values = funnel_data['Sales']
max_val = values.max()

# 3. Calculate offsets to center the bars (Funnel effect)
left_offsets = [(max_val - val) / 2 for val in values]

# 4. Plot the funnel chart
plt.barh(labels, values, left=left_offsets, color='skyblue')
plt.gca().invert_yaxis()  # Put the largest value at the top
plt.title('Sales Funnel by Outlet Tier')
plt.xlabel('Total Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Group by Year and Sum Sales
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum()

# 2. Plot the Line Chart
sales_by_year.plot(kind='line', marker='o', color='teal', figsize=(10, 5))

# 3. Add Labels
plt.title('Total Sales by Outlet Establishment Year')
plt.xlabel('Establishment Year')
plt.ylabel('Total Sales')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
